```python
import numpy as np
import matplotlib.pyplot as plt
  
# Using uniform() method
gfg = np.random.uniform(-5, 5, 5000)

print(gfg[:10])
  
plt.hist(gfg, bins = 50, density = True)
plt.show()
```

```python
# https://stepik.org/lesson/9166/step/3?discussion=3928410&reply=4931194&unit=1828
from numpy import *
from matplotlib.pyplot import *

n = 10

figure(figsize=(10, n * 3))
d = 1_000_000
heights = array([0.0] * d)

for i in range(n):
    subplot(n, 2, i+1)
    axis([0, n, 0, 1])
    hist(heights, bins=150, density=True)
    print(heights[:5])
    heights += random.uniform(low=0, high=1, size=d)
show()
```
